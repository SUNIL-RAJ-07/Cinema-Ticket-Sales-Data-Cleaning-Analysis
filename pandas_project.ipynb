Calculated Columns / Transformations(creating new columns)

Added Net_Amount

```python
df['Net_Amount'] = df['Price'] * df['Quantity'] * (1 - df['Discount'] / 100)
```

Added Booking_Lag

```python
df['Show_Date']=pd.to_datetime(df['Show_Date'])
df['Booking_Date']=pd.to_datetime(df['Booking_Date'])
df['Booking_Lag'] = (df['Show_Date'] - df['Booking_Date']).dt.days
```

Added Revenue_Category ( Low  Total_Amount < 500) (Medium  500–1500) (High  > 1500)

```python
def revenue_category(x):
    if x < 500:
        return 'Low'
    elif x <= 1500:
        return 'Medium'
    else:
        return 'High'

df['Revenue_Category'] = df['Total_Amount'].apply(revenue_category)
```

Added Customer_Type

```python
# counting how many times each Customer_ID appears
counts = df['Customer_ID'].value_counts()

# createing column using map
df['Customer_Type'] = df['Customer_ID'].map(counts)

# replaceing numbers with labels
df.loc[df['Customer_Type'] == 1, 'Customer_Type'] = 'New'
df.loc[df['Customer_Type'] != 'New', 'Customer_Type'] = 'Returning'
```

Added Is_Weekend_Show

```python
# get day number (Mon=0, Sun=6)
df['Day_Number'] = df['Show_Date'].dt.dayofweek

# default value
df['Is_Weekend_Show'] = 'No'

# set weekend as Yes
df.loc[df['Day_Number'] >= 5, 'Is_Weekend_Show'] = 'Yes'

# optional: remove helper column
df.drop(columns='Day_Number', inplace=True)
```

Encodeing Payment_Status

```python
df['Payment_Status'] = df['Payment_Status'].fillna('Pending')
```

```python
df['Payment_Status_Encoded'] = df['Payment_Status'].map({
    'Paid': 1,
    'Pending': 2,
    'Failed': 3
})
```

Standardizeing all string columns

```python
string_cols = df.select_dtypes(include='object').columns

for i in string_cols:
    df[i] = df[i].astype(str).str.strip().str.title()
```

Extracting Show_Month and Show_Year

```python
df['Show_Month'] = df['Show_Date'].dt.month
df['Show_Year'] = df['Show_Date'].dt.year
```

Converting all Date columns to YYYY-MM-DD

```python
date_cols = ['Show_Date', 'Booking_Date']

for i in date_cols:
    df[i] = df[i].dt.strftime('%Y-%m-%d')
```
